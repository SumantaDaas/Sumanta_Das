## Product Pricing (Dictioneries)

```python
# 1. Create price_set with atleast 6 entries ##
price_dict = {
    "Laptop": 1200,
    "Smartphone": 800,
    "Headphones": 150,
    "Monitor": 300,
    "Keyboard": 50,
    "Mouse": 25
}
```

```python
# 2. Code blocks for spesific operations ##
# Add a new product 
price_dict["Webcam"] = 75
```

```python
# 3. Update the price of an existing product ##
price_dict["Smartphone"] = 750  # Updated price
```

```python
# 4. Remove a product by name (handling non-existence with.pop ##
product_to_remove = "Desk Lamp"
removed_price = price_dict.pop(product_to_remove, "Product not found")
print(f"Remova status for: '{product_to_remove}, Price: {removed_price}")
```

```python
## Print average price of all products (Use basic arithmetic operations) ##
total_price = sum(price_dict.values())
average_price = total_price / len(price_dict)
print(f"Average price of all products: {average_price: .2f}")
```

```python
## Extra (optional): Print max and min priced products ##
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print(f"Max priced product: {max_product} at {price_dict[max_product]}")
print(f"Min priced product: {min_product} at {price_dict[min_product]}")
```
