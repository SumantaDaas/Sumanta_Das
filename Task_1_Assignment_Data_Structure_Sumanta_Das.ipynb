## Task 1: Product Collection (List & Tuples)

```python
## Createing a list of atleast 6 products ##
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphone", "Webcame"]
print(products)
```

```python
## Creating a tuple for another sample product list (name, price, quantity) ##
sample_product = ("Laptop", 9999.99, "Electronics")
```

```python
## Print the 2nd (index 1) and last (index -1) elements of the products list ##
print(f"2nd product: {products[1]}")
print(f"Last product: {products[-1]}")
```

```python
## Apend two new products to the products and print the updated list ##
products.append("Printer")
products.append("Scanner")
print(f"Updated products list: {products}")
```

```python
## Extra  - Convert tuple to list and change price, convert back ##
temp_list = list(sample_product)
temp_list[1] = 8999.99 # Change price
sample_product = tuple(temp_list)
print(f"Updated sample product: {sample_product}")  
```
