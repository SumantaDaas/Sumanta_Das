Process multiple orders (for loop)

```python
orders = [1200, 2500, 3200, 4500, 6000]

total_revenue = 0
discounted_orders_count = 0

print(f"{'Order Amount':<15}{'Discount (%)':<12}{'Final Total':<12}")
print("-" * 45)
```

```python
# Process each order
for amount in orders:
    # Determine discount percentage
    if amount >= 2000:
        discount_percentage = 15
    elif 1500 <= amount < 2000:
        discount_percentage = 10
    elif 1000 <= amount < 1500:
        discount_percentage = 5
    else:
        discount_percentage = 0
    
    # Calculate final amount for this order
    discount_value = amount * (discount_percentage / 100)
    final_amount = amount - discount_value
    
    # Update totals
    total_revenue += final_amount
    if discount_percentage > 0:
        discounted_orders_count += 1
        
    # Print table row
    print(f"{amount:<15}{discount_percentage:<11}{final_amount:<12.2f}")

print("-" * 45)
print(f"Total Revenue: ${total_revenue:.2f}")
print(f"Discounted Orders: {discounted_orders_count}")
```
