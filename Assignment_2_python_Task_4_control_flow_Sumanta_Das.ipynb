Loop control with conditions (break and continue)

```python
daily_sales = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for sale in daily_sales:
    # If a day's value is -1, stop processing (break)
    if sale == -1:
        print("Corrupted data found (-1). Stopping process.")
        break

    # If a day's value is 0, skip adding to revenue (continue)
    if sale == 0:
        print("No sales today. Skipping.")
        continue

    # For valid positive sales
    total_sales += sale
    print(f"Added {sale}. Running total: {total_sales}")

print(f"Final Total Sales: {total_sales}")
```
