Model Training and Evaluation - Algorithm

Types of regression model - Linear, Ridge, Lasso, Elasticnet, DecisionTree, RandomForest, XGBoost, AdaBoost, CatBoost, Kneighbour, SupportVector (SVR), Bagging, Stacking

Model_Output -> How corrct our model is: mean_square_error, mean_absolute_error, root_mean_square_error

```python
##Regressions:

#Linear Regression

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
##Evaluation metrics -> mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(Y_test, y_predict)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test, y_predict)
```

```python
from sklearn.linear_model import Ridge
r=Ridge(alpha=1.0)
```

```python
r.fit(X_train, Y_train)
```

```python
y_predict=r.predict(X_test)
```

```python
mean_absolute_error(Y_test, y_predict)
```

```python
from sklearn.linear_model import Lasso
l=Lasso(alpha=1.0)
```

```python
l.fit(X_train, Y_train)
```

```python
y_predict=l.predict(X_test)
```

```python
mean_absolute_error(Y_test, y_predict)
```

```python
from sklearn.linear_model import ElasticNet
e=ElasticNet(alpha=0.01, l1_ratio=0.5)
```

```python
e.fit(X_train, Y_train)
```

```python
y_predict=e.predict(X_test)
```

```python
mean_absolute_error(Y_test, y_predict)
```

## Decision Tree: Classification and Regression

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor(max_depth=5)
tree.fit(X_train, Y_train)
y_predict=tree.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

RandomForest Regressor: Ensemble Learning -> Combination of multiple algorith (Decision Tree)

```python
from sklearn.ensemble import RandomForestRegressor
forrest=RandomForestRegressor(n_estimators=100, random_state=42)
forrest.fit(X_train, Y_train)
y_predict=forrest.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

##In case of Ensemble learning --> XGBoost Regressor
#                              --> AdaBoost Regressor
#                              --> catBoost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor(n_estimators=100, random_state=42)
ada.fit(X_train, Y_train)
y_predict=ada.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
from catboost import CatBoostRegressor
cat=CatBoostRegressor(n_estimators=100, random_state=42, verbose=0)
cat.fit(X_train, Y_train)
y_predict=cat.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
## Kneighbours Regressor

from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, Y_train)
y_predict=knn.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
##Support Vector Regressor

from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(X_train, Y_train)
y_predict=svr.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
##Bagging regressor --> Ensemble method

from sklearn.ensemble import BaggingRegressor
bag=BaggingRegressor(n_estimators=100)
bag.fit(X_train, Y_train)   
y_predict=bag.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
##Random Forest Regressor --> Ensemble method

from sklearn.ensemble import RandomForestRegressor
forrest=RandomForestRegressor(n_estimators=100, random_state=42)
forrest.fit(X_train, Y_train)
y_predict=forrest.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
##Stacking Regressor --> Ensemble method

from sklearn.ensemble import StackingRegressor
estimators=[('rf', RandomForestRegressor(n_estimators=100, random_state=42)),
            ('xgb', XGBRegressor(n_estimators=100, random_state=42)),
            ('cat', CatBoostRegressor(n_estimators=100, random_state=42, verbose=0))]
stack=StackingRegressor(estimators=estimators, final_estimator=LinearRegression())
stack.fit(X_train, Y_train)
y_predict=stack.predict(X_test)
mean_absolute_error(Y_test, y_predict)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import BaggingRegressor

estimators=[('rf', RandomForestRegressor()),('br', BaggingRegressor())]
stack=StackingRegressor(estimators=estimators, final_estimator=DecisionTreeRegressor())
```
