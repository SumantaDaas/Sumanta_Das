User Menu (while loop + break/continue)

```python
all_orders = []

while True:
    print("\n--- Menu Options ---")
    print("1 - Add order amount to a running list")
    print("2 - show all orders and totals after applying discounts")
    print("3 - Exit")
    
    choice = input("Choose an option: ").lower()
    
    if choice == '3':
        print("Exiting the program.")
        break
    
    elif choice == '1':
        amount_input = input("Enter order amount: ")
        if amount_input.isdigit():
            all_orders.append(int(amount_input))
            print(f"Added order: ${amount_input}")
        else:
            print("Invalid amount! Please enter a number.")
        continue
    
    elif choice == '2':
        if not all_orders:
            print("No orders recorded yet!")
            continue
        print("\nOrder Summary:")
        total_revenue = 0
        for amount in all_orders:
            if amount >= 2000:
                disc = 15
            elif 1500 <= amount < 2000:
                disc = 10
            elif 1000 <= amount < 1500:
                disc = 5
            else:
                disc = 0
            
            final_amount = amount - (amount * disc / 100)
            total_revenue += final_amount
            print(f"Original: ${amount}, Discount: {disc}%, Final: ${final_amount:.2f}")
        
        print(f"Total Revenue: ${total_revenue:.2f}")
        continue
    
    else:
        print("Invalid choice. Please try again.")
```
