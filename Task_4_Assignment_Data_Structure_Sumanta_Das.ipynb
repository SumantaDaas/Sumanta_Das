## Combained operations

```python
# Assuming these variables exist from previous tasks ##

products = ["Laptop", "Smartphone", "Headphones", "Monitor", "Keyboard", "Mouse", "Webcam"]

category_list = ["Electronics", "Electronics", "Audio", "Hardware", "Hardware", "Hardware", "Electronics"]

price_dict = {"Laptop": 1200, "Smartphone": 800, "Headphones": 150, "Monitor": 300, "Keyboard": 50, "Mouse": 25, "Webcam": 75}

## 1. Create a list of tuples named 'catalog'
# We use zip to pair the product names, their prices from our dict, and categorie

catalog = []
for i in range(len(products)):
    name = products[i]
    price = price_dict.get(name, 0.0)
    category = category_list[i] if i < len(category_list) else "Uncategorized"
    catalog.append((name, price, category))
    print(f"Catalog: {catalog}")  # Print each tuple as it's added
```
