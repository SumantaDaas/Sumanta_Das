Functions (User-defined, Recursive, Lambda, Map, Filter)

```python
def factorial(n):
    #Handle negative numbers
    if n < 0:
        print("Error: Factorial is not defined for negative numbers.")
        return None
    
    # Base case: 0! and 1! are both equal to 1
    if n == 0 or n == 1:
        return 1
    
    # Recursive case: n! = n * (n-1)!
    return n * factorial(n - 1)

#Test the function
print(f"factorial(5) = {factorial(5)}")  # Should print 120
print(f"factorial(0) = {factorial(0)}")  # Should print 1
print(f"factorial(-3) = {factorial(-3)}")  # Should print an error message and return None
```

Using filter

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

#1. Prices greater than 500
expensive = list(filter(lambda p: p > 500, prices))

#2. Prices less than or equal to 500
affordable = list(filter(lambda p: p <= 500, prices))

print("Expensive (> 500):", expensive)
print("Affordable (<= 500):", affordable)
      
```

Combined Utility Function

```python
def process_prices(price_list):
    #Apply 10% discount using map + lambda
    discounted_prices = list(map(lambda p: p * 0.9, price_list))
    
    #Keep only those above 300 using filter
    filter_prices = list(filter(lambda p: p > 300, discounted_prices))
    
    return discounted_prices, filter_prices

#Test the function
disc, filt = process_prices([100, 500, 900, 50, 750])
print(f"Discounted: {disc}")
print(f"Filtered ( > 300): {filt}")
```

Mini Problem (Menu System)

```python
#Core Functions

def add_prices(prices_list, price): prices_list.append(price)

def get_average_price(prices_list):
    if not prices_list:
        return 0
    return sum(prices_list) / len(prices_list)

def get_max_price(prices_list):
    if not prices_list:
        return 0
    return max(prices_list)

#Menu Logic
my_prices = []
while True:
    print("\n--- Price Menu ---:")
    print("1. Add a price")
    print("2. Show average price")
    print("3. Show maximum price")
    print("4. Exit")
    
    choice = input("Select an option: ").lower()
    
    if choice == '1':
        val = float(input("Enter price amount: "))
        add_prices(my_prices, val)
        print(f"Price {val} added.")
    elif choice == '2':
        avg_price = get_average_price(my_prices)
        print(f"Average Price: {avg_price}")
    elif choice == '3':
        max_price = get_max_price(my_prices)
        print(f"Maximum Price: {max_price}")
    elif choice == '4':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please select from the menu options.")
```
