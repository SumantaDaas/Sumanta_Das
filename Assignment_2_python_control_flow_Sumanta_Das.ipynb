Python control flow - conditionals and loops

```python
#
# Discount Rules 

user_input = input("Enter the order amount: ")

try:
    # Validate if it is an integer and positive number
    order_amount = int(user_input)
    
    # Determine the discount percentage
    if order_amount >= 2000:
        discount_percentage = 15
    elif order_amount >= 1500:
        discount_percentage = 10
    elif order_amount >= 1000:
        discount_percentage = 5
    else:
        discount_percentage = 0

    # Calculations
    discount_value = order_amount * (discount_percentage / 100)
    subtotal = order_amount - discount_value
    tax = subtotal * 0.10
    final_total = subtotal + tax

    # Results
    print(f"\n--- Order Summary ---")
    print(f"Order Amount:     ${order_amount:,.2f}")
    print(f"Discount Applied: {discount_percentage}%")
    print(f"Tax (10%):        ${tax:,.2f}")
    print(f"Final Total:      ${final_total:,.2f}")

except ValueError:
    print("Error: Please enter a valid number.")
```
